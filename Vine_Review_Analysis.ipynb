```python
# Dependencies
import pandas as pd
```

```python
# Read our Data file with the pandas library
df = pd.read_csv("vine_table.csv")
```

```python
df.head()
```

```python
count_df = df.loc[df["total_votes"] >= 20]
count_df
```

```python
helpful_df = count_df.loc[(count_df["helpful_votes"] / count_df["total_votes"]) >= .5]
helpful_df
```

```python
paid_df = helpful_df.loc[helpful_df["vine"] == "Y"]
paid_df
```

```python
not_paid_df = helpful_df.loc[helpful_df["vine"] == "N"]
not_paid_df
```

```python
paid_total = paid_df["review_id"].count()
paid_total
```

```python
not_paid_total = not_paid_df["review_id"].count()
not_paid_total
```

```python
paid_5star = paid_df.loc[(paid_df["star_rating"] == 5.0)]
paid_5star.head(5)
```

```python
not_paid_5star = not_paid_df.loc[(not_paid_df["star_rating"] == 5.0)]
not_paid_5star.head(5)
```

```python
paid_5star_count = paid_5star["star_rating"].count()
paid_5star_count
```

```python
not_paid_5star_count = not_paid_5star["star_rating"].count()
not_paid_5star_count
```

```python
paid_5stsr_prcnt = paid_5star_count / paid_total * 100
paid_5stsr_prcnt
```

```python
not_paid_5stsr_prcnt = not_paid_5star_count / not_paid_total * 100
not_paid_5stsr_prcnt
```
